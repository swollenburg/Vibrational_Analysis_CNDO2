This notebook is used to calculate the average bond distance between different molecules. Coordinates obtained from the output can be inserted here.

```python
import math

def dist(a, b):
    return math.sqrt((a[0]-b[0])**2 + (a[1]-b[1])**2 + (a[2]-b[2])**2)
```

### CO2

```python
o1 = [-2.224459836, 6.367753376e-11, -7.323049538e-12]
c = [1.611710805e-10, -1.275578972e-10, 1.47964302e-11]
o2 = [2.224459836, 6.388036342e-11, -7.473380659e-12]
print(dist(o1, c))
print(dist(c, o2))
print()
print((dist(o1, c) + dist(c, o2))/2)
```
